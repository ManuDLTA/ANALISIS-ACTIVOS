# LIMPIEZA

```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

## FUNCIONES 

```python
def modificar_nombres_columnas(datos):

    columnas_actuales = datos.columns.tolist()
    nuevos_nombres = [unidecode(columna).upper() for columna in columnas_actuales]
    datos.columns = nuevos_nombres
    
    return datos
```

```python
def modificar_tipo_dato_columnas(datos):
    
    datos['FECHA'] = pd.to_datetime(datos['FECHA'], format='%d.%m.%Y')
    datos['ULTIMO'] = datos['ULTIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['APERTURA'] = datos['APERTURA'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['MAXIMO'] = datos['MAXIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['MINIMO'] = datos['MINIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['VOL.'] = datos['VOL.'].str.replace('\.', '', regex=True).str.replace('[^0-9]', '', regex=True).str.replace(',', '').astype(int)
    datos['% VAR.'] = datos['% VAR.'].astype(str)
    return datos
```

```python
def modificar_tipo_dato_columnas_SP500_ORO(datos):
    
    datos['FECHA'] = pd.to_datetime(datos['FECHA'], format='%d.%m.%Y')
    datos['ULTIMO'] = datos['ULTIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['APERTURA'] = datos['APERTURA'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['MAXIMO'] = datos['MAXIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['MINIMO'] = datos['MINIMO'].str.replace('\.', '', regex=True).str.replace(',', '').astype(int)
    datos['% VAR.'] = datos['% VAR.'].astype(str)
    return datos
```

### DATA BTC

```python
df_BTC = pd.read_csv("Histórico del Bitcoin (BTC) - Investing.com.csv")
display(df_BTC)
```

```python
df_BTC.info()
```

```python
df_BTC_limpio = modificar_nombres_columnas(df_BTC)
```

```python
df_BTC_limpio = modificar_tipo_dato_columnas(df_BTC)
```

```python
df_BTC_limpio.info()
```

```python
display(df_BTC_limpio)
```

```python
df_BTC_limpio.to_csv('BTC_limpio', index=False)
```
